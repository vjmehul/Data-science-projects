```python
# Study the web

url = 'https://indeed.com/Vi%E1%BB%87c-l%C3%A0m?q=data+science&l='
source_doc = requests.get(url).text
source_tree = BeautifulSoup(source_doc, 'html.parser')
```

```python
job_count_string = source_tree.find(id = 'searchCount').contents[0]
job_count_string = job_count_string.split()
job_count=job_count_string[3]
job_count = job_count[:-3].replace(",", "") + job_count[-3:]
```

## We need to select all the full time job opening in indeed.com in USA

```python
# scrapping data from aall the pages
num_pages = int(job_count)/10
num_pages=int(num_pages)

base_url = 'https://www.indeed.com'
job_links = []

for i in range(num_pages + 1):
    if i == 0:
        url = 'https://www.indeed.com/jobs?q=data+science&jt=fulltime&sort=date'
    else:
        url = 'https://www.indeed.com/jobs?q=data+science&jt=fulltime&sort=date&start=' + str(i*10)
    
    html_page = requests.get(url).text
    bs_tree = BeautifulSoup(html_page, 'html.parser')
    try:
        job_area = bs_tree.find(id = 'resultsCol')
    except:
        continue
    try:
        job_post = job_area.findAll('div')
    except:
        continue
    job_post = [jp for jp in job_post if not jp.get('class') is None
               and ''.join(jp.get('class')) == 'rowresult']
    job_ids = [jp.get('data-jk') for jp in job_post]
    
    for ids in job_ids:
        job_links.append(base_url + '/viewjob?jk=' + ids)
     
    time.sleep(1)

print('We found {} jobs'.format(len(job_links)))
```
